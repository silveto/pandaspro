```python
import pandas as pd
import numpy as np
df = pd.read_table('./../data/DF-A/DF-A_NASKNX07A.txt', header=None)
```

```python
# Get Row Index for Starting Rows
header_list = df[df[0].str.contains('kbytes')].index.values
header_list
```

```python
# Get Row Index for Ending Rows
end_list = df[df[0].str.contains('^#')].index.values
end_list
```

```python
idx_list = []
for x in header_list:
    y = end_list[np.where(x<end_list)]
    if len(y)>0:        
        print(str(x) + ' ends ' + str(y[0]))
        idx_list.append([x, y[0]]) # We are using the first number in array y e.g. y[0]
    else:
        print(str(x)+' until end of file') 
        idx_list.append([x,]) # End of file has no y[0]

print idx_list
# df[idx_list[0][0]:idx_list[0][1]] 
# [0][0] is the first element in the first list idx_list e.g. 5
# [0][1] is the second element in the first list idx_list e.g. 83

# df[idx_list[1][0]:idx_list[1][1]]
# [0][0] is the first element in the first list idx_list e.g. 88
# [0][1] is the second element in the first list idx_list e.g. 166
```

```python
counter = 0
for idxs in idx_list:
    if len(idxs)>1:
        print(idxs)
        df.loc[idxs[0]:idxs[1],'Aggregate'] = df[idxs[0]:idxs[1]][0][1:].apply(lambda x: str.split(x)[0])
        df.loc[idxs[0]:idxs[1],'kbytes'] = df[idxs[0]:idxs[1]][0][1:].apply(lambda x: str.split(x)[1])
        df.loc[idxs[0]:idxs[1],'used'] = df[idxs[0]:idxs[1]][0][1:].apply(lambda x: str.split(x)[2])
        df.loc[idxs[0]:idxs[1],'avail'] = df[idxs[0]:idxs[1]][0][1:].apply(lambda x: str.split(x)[3])
        df.loc[idxs[0]:idxs[1],'capacity'] = df[idxs[0]:idxs[1]][0][1:].apply(lambda x: str.split(x)[4])

        df.loc[idxs[0]:idxs[1],'date'] = pd.to_datetime(df[idxs[0]-2:idxs[1]-1][0].str[6:26][0:1]).dt.strftime("%d/%m/%Y %H:%M:%S").values[0]
        df.loc[idxs[0]:idxs[1],'ASUP_Type'] = df[idxs[0]-2:idxs[1]-1][0].str.extract('\((.*?)\)', expand=True)[0].values[0]
        df.loc[idxs[0]:idxs[1],'hostname'] = df[idxs[0]-1:idxs[1]][0].str.split().values[0][2]
        df.loc[idxs[0]:idxs[1],'serial_no'] = df[idxs[0]-1:idxs[1]][0].str.split().values[0][4]
    else:
        print(idxs)
        df.loc[idxs[0]:,'Aggregate'] = df[idxs[0]:][0][1:].apply(lambda x: str.split(x)[0])
        df.loc[idxs[0]:,'kbytes'] = df[idxs[0]:][0][1:].apply(lambda x: str.split(x)[1])
        df.loc[idxs[0]:,'used'] = df[idxs[0]:][0][1:].apply(lambda x: str.split(x)[2])
        df.loc[idxs[0]:,'avail'] = df[idxs[0]:][0][1:].apply(lambda x: str.split(x)[3])
        df.loc[idxs[0]:,'capacity'] = df[idxs[0]:][0][1:].apply(lambda x: str.split(x)[4])

        df.loc[idxs[0]:,'date'] = pd.to_datetime(df[idxs[0]-2:-1][0].str[6:26][0:1]).dt.strftime("%d/%m/%Y %H:%M:%S").values[0]
        df.loc[idxs[0]:,'ASUP_Type'] = df[idxs[0]-2:-1][0].str.extract('\((.*?)\)', expand=True)[0].values[0]
        df.loc[idxs[0]:,'hostname'] = df[idxs[0]-1:][0].str.split().values[0][2]
        df.loc[idxs[0]:,'serial_no'] = df[idxs[0]-1:][0].str.split().values[0][4]
```

```python
df.drop([0], axis=1, inplace=True)
```

```python
result_df = df[~df['Aggregate'].isnull()].reset_index(drop=True)
result_df = result_df[['date', 'ASUP_Type', 'hostname', 'serial_no', 'Aggregate', 'kbytes', 'used', 'avail', 'capacity']]
```

```python
active_result_df = result_df[result_df['Aggregate'].str.contains(".snapshot") == False]
active_result_df = active_result_df.reset_index(drop=True)
active_result_df.columns = ['date', 'ASUP_Type', 'hostname', 'serial_no', 'active_Aggregate', 'aggr_kbytes', 'aggr_used', 'aggr_avail', 'aggr_capacity']    
active_result_df
```

```python
snapshot_result_df = result_df[result_df['Aggregate'].str.contains(".snapshot")]
snapshot_result_df = snapshot_result_df.reset_index(drop=True)
snapshot_result_df.columns = ['date', 'ASUP_Type', 'hostname', 'serial_no', 'snap_Aggregate', 'snap_kbytes', 'snap_used', 'snap_avail', 'snap_capacity']
snapshot_result_df.drop(['date', 'ASUP_Type', 'hostname', 'serial_no'], axis=1, inplace=True)
snapshot_result_df
```

```python
output_result_df = pd.concat([active_result_df, snapshot_result_df], axis=1)
output_result_df
```
